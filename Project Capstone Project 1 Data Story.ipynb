## 1. Exploratory data analysis
First, I will make histogram and scatterplot. I want to identify which features corellate best with prices. 

```python
full_list = ['MSSubClass' , 'LotArea' , 'OverallQual' , 'OverallCond' , 'YearBuilt' , 'YearRemodAdd' , 'BsmtFinSF1' ,
           'BsmtFinSF2' , 'BsmtUnfSF' , 'TotalBsmtSF' , '1stFlrSF' , '2ndFlrSF' ,'LowQualFinSF', 'GrLivArea','BsmtFullBath',
           'BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars',
           'GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold',
           'YrSold', 'SalePrice']

for i in full_list:
    plt.hist(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
for i in full_list:
    plt.scatter(df[i],df['SalePrice'])
    plt.xlabel(i)
    plt.ylabel('Sale Price')
    plt.show()
    print (np.corrcoef(df[i], df['SalePrice']))
```

# 2. Hypothesis
Based on the scatter plots and corellation efficiency, the observation is that sale prices corellate best with:
1. OverallQual: Overall material and finish quality (R square =0.791)
2. GrLivArea: Above grade (ground) living area square feet (R square = 0.708)
3. GarageArea: Size of garage in square feet (R square = 0.623)
4. GarageCars: Size of garage in car capacity (R square = 0.640)
5. TotRmsAbvGrd: Total rooms above grade (does not include bathrooms)(R square = 0.534)

Therefore, I hypothsize that sale price of a house can be predicted using data from the above features.
